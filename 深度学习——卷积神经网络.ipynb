```python
import glob 
import PIL
import PIL.Image as Image
import numpy as np
import random
import matplotlib.pyplot as plt

import tensorflow 

from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
 
def 数据处理(s):                    #pathw为需要处理数据的文件地址
    lable,testx,lableY=[],[],[]
    tem_file1=glob.glob(s+'\\'+'*')
    for each in tem_file1:
        lable.append(str(each)[-4:])
    for t in lable:
        tem_pic1=glob.glob(s+'\\'+str(t)+'\\'+'**.bmp')
        
        for file in tem_pic1:
            image = Image.open(file)
            v_img=np.array(image).ravel()
            testx.append(v_img)                          #添加图片到数组
            lableY.append(str(t))                        #添加标签
    return testx,lableY



if __name__=="__main__":
    
    tranX,lableX=数据处理(r'F:\图像识别数据\No4\测试样本')     
    testX,lableC=数据处理(r'F:\图像识别数据\No4\训练样本')
    
   
    print(len(tranX))
    print(len(testX))
    
    print(len(lableX))
    print(len(lableC))
    
    (tranX, lableX), (testX, lableC) = mnist.load_data()
    x_Train4D=tranX.reshape(tranX.shape[0],28,28,1).astype('float32')#将图像特征值转换为四维矩阵
    x_Test4D=testX.reshape(testX.shape[0],28,28,1).astype('float32')
    
    x_Train4D_normalize = x_Train4D / 255#将图像特征值（features）标准化可以提高模型预测的准确度，并且更快收敛
    x_Test4D_normalize = x_Test4D / 255
    
    y_TrainOneHot = np_utils.to_categorical(lableX)#将训练数据和测试数据的label进行one-hot encoding转换
    y_TestOneHot = np_utils.to_categorical(lableC)
    
    
    
    
    model = Sequential()
    
    model.add(Conv2D(filters=16,kernel_size=(5,5),padding='same',activation='relu',input_shape=(28,28,1)))
    
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(filters=16,kernel_size=(5,5),padding='same',activation='relu',input_shape=(28,28,1)))  
    model.add(MaxPooling2D(pool_size=(2,2)))
  
    model.add(Flatten())
    
    #bp网络
    model.add(Dense(128, activation='relu'))
  
    model.add(Dense(10,activation='softmax'))
    print(model.summary())
    
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    train_history=model.fit(x=x_Train4D_normalize, y=y_TrainOneHot,validation_split=0.2,epochs=20, batch_size=300,verbose=2)
    prediction=model.predict_classes(x_Test4D_normalize)
    sum=0
    for i in range(0,len(prediction)):
        if(prediction[i]==lableC[i]):
            sum=sum+1    
    #y.append(sum/len(lableC))
    print(sum/len(lableC))

    
```
