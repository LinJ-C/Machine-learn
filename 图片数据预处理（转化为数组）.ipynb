```python
import glob 
import PIL
import PIL.Image as Image
import numpy as np
from sklearn import tree
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
 
def 数据处理(s):                    #pathw为需要处理数据的文件地址
    lable,testx,lableY=[],[],[]
    tem_file1=glob.glob(s+'\\'+'*')
    for each in tem_file1:
        lable.append(str(each)[-4:])
    for t in lable:
        tem_pic1=glob.glob(s+'\\'+str(t)+'\\'+'**.bmp')
        
        for file in tem_pic1:
            image = Image.open(file)
            v_img=np.array(image).ravel()
            testx.append(v_img)                          #添加图片到数组
            lableY.append(str(t))                        #添加标签
    return testx,lableY

#四个参数，n降维梯度，x数据集，lx训练集标签，lc测试集标签
def 预测结果(n,x,lablex,lablec):                         #以多少维度阶梯降维
    y=[]
    for i in range(1,10):
        shu=i*n                                           #设定的维度
        print(shu)
        pca=PCA(n_components=shu)        
        newX = pca.fit_transform(x)
                                                          #获得降维后的数据       
        a=newX[:1342]                   
        b=newX[1342:]
        
    #使用决策数方法，训练
        clf = tree.DecisionTreeClassifier() 
    #用测试集的数据和与之对应的标签进行训练  
        clf = clf.fit(a,lablex)
    #用模型预测
        get=clf.predict(b)
    #求准确率   
        sum=0
        for i in range(0,len(get)):
            if(get[i]==lablec[i]):
                sum=sum+1
        y.append(sum/len(get))
        print(sum/len(get))
    return y

if __name__=="__main__":
    
    tranX,lableX=数据处理(r'F:\图像识别数据\No4\测试样本')
        
    testX,lableC=数据处理(r'F:\图像识别数据\No4\训练样本')
    



    print(len(lableC))
    print(len(lableX))

    print(len(tranX))
    
    print(len(testX))

    data=tranX+testX
    X=np.array(data)
    print(len(data))

    y=np.array(预测结果(5,X,lableX,lableC))
    
    plt.plot(y)
    plt.show()
```
