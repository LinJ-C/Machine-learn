```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(100000)
y=np.random.rand(100000)
o=[]
q=[]
a=[]
b=[]
cum=0
for i in range(0,100000):
    get=x[i]*x[i]+y[i]*y[i]
    if(get<1):
        o.append(x[i])
        q.append(y[i])
        cum=cum+1
    else:
        a.append(x[i])
        b.append(y[i])
plt.scatter(o,q,c='b')
plt.scatter(a,b,c='r')
plt.show()
print(4*cum/100000)
```
